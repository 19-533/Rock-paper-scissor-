```python
import random
rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

list=[]
list.append(rock)
list.append(paper)
list.append(scissors)
a=int (input("what do you choose? Type 0 for Rock, 1 for Paper or 2 for Scissors" ))

if a == 0:
  print(rock)
  b=random.randint(0,len(list)-1)
  c=list[b]
  print("Computer chose")
  if c == list[0]:
    print("It's a Draw")
  elif c == list[1]:
    print("You lose")
  else:
    print("You win")
elif a == 1:
  print(paper)
  b=random.randint(0,len(list)-1)
  c=list[b]
  print("Computer chose"+c)
  if c == list[1]:
    print("It's a Draw")
  elif c == list[2]:
    print("You lose")
  else:
    print("You win")
elif a == 2:
  print(scissors)
  b=random.randint(0,len(list)-1)
  c=list[b]
  print("Computer chose"+c)
  if c == list[2]:
    print("It's a Draw")
  elif c == list[0]:
    print("You lose")
  else:
    print("You win")



```
